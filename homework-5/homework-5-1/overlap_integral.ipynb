```python
plt.plot(monte_carlo.n_list, monte_carlo.important_sample_overlap)
plt.axhline(0.734, color="black", linestyle="--")
plt.xlabel('Number of points sampled')
plt.ylabel('Integral Value')
plt.xscale('log')
plt.title("Overlap integral of two H atom 2pz orbitals (Importance Sampling)")
plt.show()
```

## Random vs Importance Sampling

The importance sampling method converges to 0.734 , while the random sampling method converges to 0.736 . These values are very similar, so we can have some confidence that the methods are accurate. Furthermore, the importance sampling method converges by a factor of 10 quicker than the random sampling method.

```python
plt.plot(monte_carlo.sep_dist, monte_carlo.overlap_vs_sep_dist)
plt.xlabel('Seperation distance $(a_{0})$')
plt.ylabel('Integral Value')
plt.axhline(0, color="black", linestyle="--")
plt.title("Overlap integral vs Seperation Distance")
plt.show()
```

## Overlap Integral as a Function of Seperation Distance
 
The change in sign is a result of the overlap between opposite phases of the two 2pz orbitals. At R = 0, the 2pz orbtials are perfectely aligned, which means the same phases are overlapping. As these orbitals move apart, the opposite phases are perfectely aligned. This causes them to destructively interfere, which is what the negative overlap integral is showing. As the orbitals are pulled apart to infinity, they are no longer in contact and have 0 overlap.

## The Efficiency of Importance Sampling

Importance sampling improves efficiency of numerical integrations by considering only the indepedent variables that have a considerable effect on the integral. This increases the computational efficiency because the number of computations is decreased while still maintaining a good approximation of the integral.

## The Significance of the Overlap Integral and its Dependence on R

The overlap integral is important because it explains the bonding behavior of molecules. When the overlap integral between two atoms is large, the atoms are capable of forming strong bonds. When the overlap integral between two atoms is negative, this suggests the formation of antibonding orbitals. When the overlap integral between two atoms is 0, this means the atoms are too far apart to engage in chemical bonding.
